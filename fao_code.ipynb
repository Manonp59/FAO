```python
df_gp_area_year = df.groupby(by=['area','year']).agg({'food_supply_(kcal/capita/day)':'sum','protein_supply_quantity_(g/capita/day)':'sum','pop_1000_hab':'mean'}).reset_index()
```

### Question 10 - Calculez le ratio énergie/poids de chaque produit et pays. Vous devriez vous apercevoir qu’étonnement, ces informations varient en fonction du pays. Pour pallier ce problème, calculez la moyenne de ce ratio pour chaque aliment.  Attention à bien gérer les valeurs égales à 0. Vérifiez la cohérence de votre calcul en comparant le résultat avec l’apport calorique d’un œuf.


```python
df_gp_area_item = df.groupby(by=['area','item']).agg({'food_supply_(kcal/capita/day)':'sum','food_supply_quantity_(kg/capita/yr)':'sum','pop_1000_hab':'mean'}).reset_index()
```

```python
df_gp_area_item[['food_supply_(kcal/capita/day)', 'food_supply_quantity_(kg/capita/yr)']] = df_gp_area_item.groupby('item')[['food_supply_(kcal/capita/day)', 'food_supply_quantity_(kg/capita/yr)']].transform(lambda x: x.replace(0, x.mean()))
```

```python
df_gp_area_item['food_supply_(kcal/yr)'] = df_gp_area_item['food_supply_(kcal/capita/day)']*365*(df_gp_area_item['pop_1000_hab']*1000)
df_gp_area_item['food_supply_quantity(kg/yr)'] = df_gp_area_item['food_supply_quantity_(kg/capita/yr)']*(df_gp_area_item['pop_1000_hab']*1000)
df_gp_area_item['ratio_energie_poids']=df_gp_area_item['food_supply_(kcal/yr)']/df_gp_area_item['food_supply_quantity(kg/yr)']
```

```python
df_ratio = df.groupby(by='item').agg({'food_supply_(kcal/capita/day)':'mean','food_supply_quantity_(kg/capita/yr)':'mean','pop_1000_hab':'mean'}).reset_index()
df_ratio['food_supply_(kcal/yr)'] = df_ratio['food_supply_(kcal/capita/day)']*365*(df_ratio['pop_1000_hab']*1000)
df_ratio['food_supply_quantity(kg/yr)'] = df_ratio['food_supply_quantity_(kg/capita/yr)']*(df_ratio['pop_1000_hab']*1000)
df_ratio['ratio_energie_poids']=df_ratio['food_supply_(kcal/yr)']/df_ratio['food_supply_quantity(kg/yr)']
```

```python
df_question_10 = df_ratio[['item','ratio_energie_poids']]
df_question_10
```
